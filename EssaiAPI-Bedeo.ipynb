```python
def collect_ue(annee):
    for ue in annee["schema"]:
        if "code" in ue:
            ues.append(ue["code"])
        else:
            for u in ue["choix"]:
                ues.append(u["code"])
```

### Export yaml pour Jekyll

```python
master = {}
master["m1"] = {}
master["m2"] = {}
```

```python
ue_raw = {}
for ue in ues:
    response = requests.get(url+ue, headers=headers)
    ue_raw[ue] = response.json()
```

```python
# ue_filtered = {key: ue_raw[key] for key in ue_raw.keys()
#        & {'code', 'code_suivi', 'affichable'}}
ue_filtered = dict(ue_raw)
```

```python
clefs = []
for ue in ue_raw:
    for k in list(ue_raw[ue].keys()):
        clefs.append(k)
set(clefs)
```

```python
champs = ['responsable','secretariat','credits','code','intitule','url']
unwanted = set(clefs) - set(champs)
for ue in ue_filtered:
    for unwanted_key in unwanted:
        if unwanted_key in ue_filtered[ue]:
            del ue_filtered[ue][unwanted_key]
```

```python
ue_categ = {
'NFE115': "bdd-si",
'RSX103': "rsxmobile",
'RCP104': "optim-ia",
'NFP103': "bdd-si",
'NFP108': "bdd-si",
'STA101': "maths",
'NFP106': "optim-ia",
'ANG330': "autre",
'RCP208': "optim-ia",
'CSC109': "maths",
'MVA107': "optim-ia",
'STA002': "maths",
'RCP106': "optim-ia",
'RCP209': "optim-ia",
'RCP211': "optim-ia",
'RCP217': "optim-ia",
'CSC217': "maths",
'SMB214': "rsxmobile",
'MTR105': "rsxmobile",
'STA211': "maths",
'STA212': "maths",
'NFE211': "bdd-si",
'NFE212': "bdd-si",
'UA332U': "autre",
}
```

```python
for ue in ues:
    if ue in ue_m1:
        master["m1"][ue] = ue_filtered[ue]
        master["m1"][ue]["categ"] = ue_categ[ue]
    elif ue in ue_m2:
        master["m2"][ue] = ue_filtered[ue]
        master["m2"][ue]["categ"] = ue_categ[ue]
#master["m2"]
```
